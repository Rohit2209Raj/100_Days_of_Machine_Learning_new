```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Titanic.csv')
df=df.drop(columns=['Unnamed: 0','who','adult_male','deck','embark_town','alive','alone'])
```

```python
num_cols=['pclass','age','sibsp','parch']
ord_cols=['class']
cat_cols=['sex','embarked']
```

# Ab kaise proceed karna hai wo dekhte hai

## Sabse pehle hame kis col pe kya karna hai wo dekhte hai
### num_cols
---> inpe sabse pehle imputer and then standardisation
### ord_cols
---> inpe only ordinal encoder
### cat_cols
---> inpe onehot encoder

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
num_pipeline=Pipeline([
    ('imputer',SimpleImputer()),
    ('scalar',StandardScaler())
])
```

```python
ord_pipeline=Pipeline([
    ('ord',OrdinalEncoder(categories=[['Third','Second','First']])),
])
```

```python
cat_pipeline=Pipeline([
    ('Onehot',OneHotEncoder(drop='first',sparse_output=False,dtype=int))
])
```

```python
transformer=ColumnTransformer(
    transformers=[
        ('tnf1',num_pipeline,num_cols),
        ('tnf2',ord_pipeline,ord_cols),
        ('tnf3',cat_pipeline,cat_cols)
    ]
    ,remainder='passthrough'
)
```

```python
pipe=Pipeline([
    ('preprocessing',transformer),
    ('predict',LogisticRegression())
])
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['survived']),df['survived'],test_size=0.25)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
acc=accuracy_score(y_test,y_pred)
acc
```
