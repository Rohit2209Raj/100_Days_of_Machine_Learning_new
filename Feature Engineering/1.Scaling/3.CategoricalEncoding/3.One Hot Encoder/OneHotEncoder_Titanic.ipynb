```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
x.columns
```

```python
num=['pclass','age','sibsp','parch','fare']
ord_col=['class']
cat=['sex']
```

```python
one=OneHotEncoder(sparse_output=False,drop='first',dtype='int')
od=OrdinalEncoder(categories=[['Third','Second','First']])
scalar=StandardScaler()
```

```python
encoded1_x_train=pd.DataFrame(one.fit_transform(x_train[cat]),columns=one.get_feature_names_out(cat), index=x_train.index)
```

```python
encoded1_x_test=pd.DataFrame(one.transform(x_test[cat]),columns=one.get_feature_names_out(cat), index=x_test.index)
```

```python
encoded2_x_train=pd.DataFrame(od.fit_transform(x_train[ord_col]),columns=ord_col,index=x_train.index).astype(int)
```

```python
encoded2_x_test=pd.DataFrame(od.transform(x_test[ord_col]),columns=ord_col,index=x_test.index).astype(int)
```

```python
encoded3_x_train=pd.DataFrame(scalar.fit_transform(x_train[num]),columns=num,index=x_train.index).astype(int)
```

```python
encoded3_x_test=pd.DataFrame(scalar.transform(x_test[num]),columns=num,index=x_test.index).astype(int)
```

```python
x_train=x_train.drop(columns=x_train.columns)
```
